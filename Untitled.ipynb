```python
def generateSample(m_x,sig_x,m_y,sig_y,m_z,sig_z,df):
    X = np.random.normal(m_x, sig_x, 300)
    Y = np.random.normal(m_y, sig_y, 300)
    Z = np.random.normal(m_z, sig_z, 300)

    for i in range(0,300):
        df.loc[i] = [X[i],Y[i],Z[i]]
```
